```python
math=dataset["Math"].values
read=dataset["Reading"].values
write=dataset["Writing"].values
```

## Using Scikit-Learn Machine Learning Approach

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
X=np.array([math,read]).T
Y=np.array(write)
```

```python
linear=LinearRegression()
linear.fit(X,Y)
Y_pred=linear.predict(X)
```

```python
rmse=np.sqrt(mean_squared_error(Y,Y_pred))
r2=linear.score(X,Y)
```

```python
print("The RMSE Score of the Model is %r" %rmse)
print("The R2 Score of the Model is %r" %r2)
```
